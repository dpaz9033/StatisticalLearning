Método IQR: Se calculan los valores atípicos (outliers) en el DataFrame 'data_clean' utilizando el método del rango intercuartílico (IQR). Para cada columna en 'data_clean', se calculan el primer cuartil (Q1) y el tercer cuartil (Q3), y luego se determinan los umbrales inferior y superior para identificar valores atípicos. Finalmente, se crea una columna adicional para cada variable que indica si un valor es un valor atípico o no.

```python
temp = pd.DataFrame()

for col in data_clean.columns:
    column = data_clean[col]

    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1

    umbral_inferior = Q1 - 1.5 * IQR
    umbral_superior = Q3 + 1.5 * IQR

    data_clean[col + '_outliers'] = (column < umbral_inferior) | (column > umbral_superior)

    #temp[col + '_outliers'] = ((column < umbral_inferior) | (column > umbral_superior))

    print(f'{Q1 =} {Q3=} {IQR =} {umbral_inferior = } {umbral_superior =}')
```

##Preparación de Cluster

```python
from sklearn.cluster import KMeans
import warnings
```

##Función de Inercia

```python
inercia = []
ks = []

for k in range(1,15):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data_clean)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

```python
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
```

```python
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

##Modelo para K Means

```python
modelo = KMeans(n_clusters=3,random_state = 10)
modelo.fit(data_clean)
```

Se agrega una nueva columna llamada cluster, que indica a qué cluster pertenece cada fila de datos, lo que facilita el análisis y la segmentación de los datos en función de los clusters identificados por el modelo de clustering.

```python
data_clean['cluster'] = modelo.labels_
data_clean
```

```python
data_clean.groupby(by='cluster').count()
```
